```python
import os
import csv
import random
import pandas as pd
import numpy as np
```

```python
root_dir = "/content/drive/MyDrive/Ulcer Dataset"
csv_file = "image_labels.csv"
```

```python
data = []
for root, dirs, files in os.walk(root_dir):
  label = os.path.basename(root)
  for file in files:
    if file.lower().endswith((".jpg", ".png", ".jpeg")):
      path = os.path.join(root, file)
      data.append([path, label])

random.shuffle(data)

train_split = int(0.8*len(data))
val_split = int(0.9*len(data))
```

```python
cleaned = []
for i, row in enumerate(data):
  if i < train_split:
    split = "train"
  elif i < val_split:
    split = "val"
  else:
    split = "test"
  cleaned.append([row[0], row[1], split])

data = cleaned
```

```python
with open(csv_file, "w", newline="") as f:
  writer = csv.writer(f)
  writer.writerow(["image_path", "label", "split"])
  writer.writerows(data)
print("CSV created ",{csv_file}," with ",{len(data)}, "rows")
```

```python
def build_file_list(root_dir, test_size=0.2):
    paths, labels, splits = [], [], []

    for label in os.listdir(root_dir):
        label_path = os.path.join(root_dir, label)
        if not os.path.isdir(label_path):
            continue
        imgs = [os.path.join(label_path, f) for f in os.listdir(label_path) if f.lower().endswith((".jpg",".png",".jpeg"))]

        train_imgs, val_imgs = train_test_split(imgs, test_size=test_size, random_state=42)

        paths.extend(train_imgs)
        labels.extend([label]*len(train_imgs))
        splits.extend(["train"]*len(train_imgs))

        paths.extend(val_imgs)
        labels.extend([label]*len(val_imgs))
        splits.extend(["val"]*len(val_imgs))

    return paths, labels, splits
```

```python
import pandas as pd
import random
import os
from sklearn.model_selection import train_test_split


df = pd.read_csv("/content/ulcer_dataset_rebuilt.csv")
df = df.reset_index(drop=True)


def biased_label(label):
    if label.lower() == 'arterial ulcer':
        age = random.randint(50, 75) if random.random() < 0.8 else random.randint(40, 85)
        sex = random.choices(["M", "F"], weights=[0.7, 0.3])[0]
        smoking = random.choices([1, 0], weights=[0.9, 0.1])[0]
        meds = random.choices(["ac", "oha", "insul", "bp", "hf", "none"],
                              weights=[0.05, 0.05, 0.25, 0.45, 0.1, 0.1])[0]
        htn = random.choices([1, 0], weights=[0.8, 0.2])[0]
        dm = random.choices([1, 0], weights=[0.75, 0.25])[0]
        location = random.choices(
            ["med_leg", "lat_leg", "dorsum", "lat_foot", "plantar", "heel", "ball", "digit"],
            weights=[0.06, 0.15, 0.15, 0.15, 0.10, 0.08, 0.04, 0.27])[0]
        periwound = random.choices(["pigm", "callus", "pale"], weights=[0.1, 0.03, 0.87])[0]
        pain_score = random.randint(6, 8) if random.random() < 0.8 else random.randint(1, 10)
        periph_pulse_absent = random.choices(["pop", "ata", "pta", "dp", "none"],
                                             weights=[0.15, 0.2, 0.2, 0.35, 0.1])[0]
        neuropathy = random.choices([1, 0], weights=[0.2, 0.8])[0]
        abi = round(random.uniform(0.4, 0.8), 2)
        hba1c = round(random.uniform(4, 10), 1)
        doppler = random.choices(["art_sten", "ven_ins", "normal", "na"],
                                 weights=[0.8, 0.05, 0.05, 0.1])[0]

    elif label.lower() == "venous ulcer":
        age = random.randint(45, 70)
        sex = random.choices(["M", "F"], weights=[0.4, 0.6])[0]
        smoking = random.choices([1, 0], weights=[0.3, 0.7])[0]
        meds = random.choices(["bp", "ac", "none"], weights=[0.5, 0.2, 0.3])[0]
        htn = random.choices([1, 0], weights=[0.6, 0.4])[0]
        dm = random.choices([1, 0], weights=[0.2, 0.8])[0]
        location = random.choices(["med_leg", "lat_leg", "ankle"], weights=[0.7, 0.2, 0.1])[0]
        periwound = random.choices(["eczema", "pigm", "lipoderm"], weights=[0.4, 0.4, 0.2])[0]
        pain_score = random.randint(2, 6)
        periph_pulse_absent = "none"
        neuropathy = 0
        abi = round(random.uniform(0.8, 1.2), 2)
        hba1c = round(random.uniform(4.5, 8), 1)
        doppler = random.choices(["ven_ins", "normal", "na"], weights=[0.8, 0.15, 0.05])[0]

    elif label.lower() == "neurotrophic ulcer":
        age = random.randint(45, 80)
        sex = random.choices(["M", "F"], weights=[0.6, 0.4])[0]
        smoking = random.choices([1, 0], weights=[0.4, 0.6])[0]
        meds = random.choices(["oha", "insul", "bp", "none"], weights=[0.3, 0.4, 0.2, 0.1])[0]
        htn = random.choices([1, 0], weights=[0.5, 0.5])[0]
        dm = 1
        location = random.choices(["plantar", "digit", "heel", "ball"], weights=[0.5, 0.2, 0.2, 0.1])[0]
        periwound = random.choices(["callus", "pigm"], weights=[0.6, 0.4])[0]
        pain_score = random.randint(0, 3)
        periph_pulse_absent = random.choices(["pta", "dp", "none"], weights=[0.3, 0.3, 0.4])[0]
        neuropathy = random.choices([1, 0], weights=[0.8, 0.2])[0]
        abi = round(random.uniform(0.6, 1.0), 2)
        hba1c = round(random.uniform(6, 12), 1)
        doppler = random.choices(["art_sten", "normal", "na"], weights=[0.6, 0.2, 0.2])[0]

    else:  # fallback for any other label
        age, sex, smoking, meds, htn, dm, location, periwound = 60, "M", 0, "none", 0, 0, "na", "na"
        pain_score, periph_pulse_absent, neuropathy, abi, hba1c, doppler = 0, "na", 0, 1.0, 5.0, "na"

    return (age, sex, smoking, meds, htn, dm, location, periwound,
            pain_score, periph_pulse_absent, neuropathy, abi, hba1c, doppler)



def build_dataset(df, outfile="dataset.csv"):
    patient_ids, ages, sexes, smokings, meds_list, htns, dms = [], [], [], [], [], [], []
    locations, periws, pain_scores, periphs, neuros, abis, hba1cs, dopplers = [], [], [], [], [], [], [], []

    for i in range(len(df)):
        label = df.loc[i, "label"].lower()   # safe row indexing
        pid = f"P{i+1:05d}"
        patient_ids.append(pid)

        (_age, _sex, _smoking, _meds, _htn, _dm, _location, _periwound,
        _pain_score, _periph_pulse_absent, _neuropathy, _abi, _hba1c, _doppler) = biased_label(label)

        ages.append(_age)
        sexes.append(_sex)
        smokings.append(_smoking)
        meds_list.append(_meds)
        htns.append(_htn)
        dms.append(_dm)
        locations.append(_location)
        periws.append(_periwound)
        pain_scores.append(_pain_score)
        periphs.append(_periph_pulse_absent)
        neuros.append(_neuropathy)
        abis.append(_abi)
        hba1cs.append(_hba1c)
        dopplers.append(_doppler)


    df["patient_id"] = patient_ids
    df["age"] = ages
    df["sex"] = sexes
    df["smoking"] = smokings
    df["meds"] = meds_list
    df["htn"] = htns
    df["dm"] = dms
    df["location"] = locations
    df["periwound"] = periws
    df["pain_score"] = pain_scores
    df["periph_pulse_absent"] = periphs
    df["neuropathy"] = neuros
    df["abi"] = abis
    df["hba1c"] = hba1cs
    df["doppler"] = dopplers


    df.to_csv(outfile, index=False)
    return df

root_dir = "/content/drive/MyDrive/Ulcer Dataset"
df = build_dataset(df, "ulcer_dataset_final.csv")
df.head(100)
```
